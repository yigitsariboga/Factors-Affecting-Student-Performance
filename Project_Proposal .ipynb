## Data Set Introduction

The dataset is a compilation of information about students academic success in mathematics. A wide range of characteristics, including demographic data, study habits, and social factors that could affect student performance, are included in the dataset. The data, which includes student demographic, social, and academic data, was gathered from two secondary schools in Portugal.

```python
import pandas as pd

data = pd.read_csv('datasets\\student-mat.csv', header = 0, sep=';')
print(data.info()) # gathering info about data set in this cell.
data
```

The dataset contains 33 variables, including the following:

1 - school : student's school (binary: "GP" - Gabriel Pereira or "MS" - Mousinho da Silveira)

2 - sex : student's gender (binary: "F" - female or "M" - male)

3 - age : student's age (numeric: from 15 to 22)

4 - address : student's home address type (binary: "U" - urban or "R" - rural)

5 - famsize : family size (binary: "LE3" - less or equal to 3 or "GT3" - greater than 3)

6 - Pstatus : parent's cohabitation status (binary: "T" - living together or "A" - apart)

7 - Medu : mother's education (numeric: 0 - none,  1 - primary education (4th grade), 2 – 5th to 9th grade, 3 – secondary education or 4 – higher education)

8 - Fedu : father's education (numeric: 0 - none,  1 - primary education (4th grade), 2 – 5th to 9th grade, 3 – secondary education or 4 – higher education)

9 - Mjob : mother's job (nominal: "teacher", "health" care related, civil "services" (e.g. administrative or police), "at_home" or "other")

10 - Fjob : father's job (nominal: "teacher", "health" care related, civil "services" (e.g. administrative or police), "at_home" or "other")

11 - reason : reason to choose this school (nominal: close to "home", school "reputation", "course" preference or "other")

12 - guardian : student's guardian (nominal: "mother", "father" or "other")

13 - traveltime :home to school travel time (numeric: 1 - <15 min., 2 - 15 to 30 min., 3 - 30 min. to 1 hour, or 4 - >1 hour)

14 - studytime : weekly study time (numeric: 1 - <2 hours, 2 - 2 to 5 hours, 3 - 5 to 10 hours, or 4 - >10 hours)

15 - failures : number of past class failures (numeric: n if 1<=n<3, else 4)

16 - schoolsup : extra educational support (binary: yes or no)

17 - famsup : family educational support (binary: yes or no)

18 - paid : extra paid classes within the course subject (Math) (binary: yes or no)

19 - activities : extra-curricular activities (binary: yes or no)

20 - nursery : attended nursery school (binary: yes or no)

21 - higher : wants to take higher education (binary: yes or no)

22 - internet : Internet access at home (binary: yes or no)

23 - romantic : with a romantic relationship (binary: yes or no)

24 - famrel : quality of family relationships (numeric: from 1 - very bad to 5 - excellent)

25 - freetime : free time after school (numeric: from 1 - very low to 5 - very high)

26 - goout : going out with friends (numeric: from 1 - very low to 5 - very high)

27 - Dalc : workday alcohol consumption (numeric: from 1 - very low to 5 - very high)

28 - Walc : weekend alcohol consumption (numeric: from 1 - very low to 5 - very high)

29 - health : current health status (numeric: from 1 - very bad to 5 - very good)

30 - absences : number of school absences (numeric: from 0 to 93)

these are grades related with the course subject:

31 - G1 : first period grade (numeric: from 0 to 20)

32 - G2 : second period grade (numeric: from 0 to 20)

33 - G3 : final grade (numeric: from 0 to 20, output target)

- Grades (G3) will be our target variable.

#### Last 3 columns names will be changed for easier understanding and be compatible with syntax. (Other columns names are normal.)

```python
data.rename(columns={'G1': 'FirstGrade', 'G2': 'SecondGrade', 'G3': 'FinalGrade'}, inplace=True)
```

#### We are checking if the data has missing values. (Determining that we need imputation or not for further actions.)

```python
check = 0
for col in data.columns:
    missing_values = data[col].isna().sum()
    if missing_values > 0:
        print(f"Column '{col}' contains {missing_values} missing values.")
        check += 1

if check == 0:
    print("Columns do not contain missing values.") # do not need imputation for now
else:
    print("Columns contain missing values.")
```

#### We are checking unique variables for consistency.

```python
unique_values = 0
for col in data.columns:
    unique_values = data[col].unique()
    print(f"Column '{col}' contains {unique_values} unique values.")
    unique_values = 0
```

We can say that our columns are consistent.

## Description of the Problem

## Methodologies

- Regression analysis: This can be use for determining the relationships between multiple variables, including socioeconomic status, study habits, and academic achievement.

- Cluster analysis: This can be used to group students based on demographic information, study habits, and social factors to determine which factors are most important for student success.

- Machine learning algorithms: These can be used to predict student academic performance based on their demographic information and study habits.

- Data visualization techniques: These can be used to explore the relationships between variables and to identify patterns or trends in the data.

- Feature selection: This method can be used to identify the most important variables to predict academic performance and to remove less important variables to simplify the model.

- Multicollinearity: This can be used when the predictor variables in the regression model are highly correlated, making it difficult to determine the independent effects of each variable. To address the Multicolinearity regarding student characteristics, techniques such as variance inflation factor (VIF) analysis can be used to identify and eliminate highly correlated variables.

- Encoding:
    - a. Categorical variables can be encoded using techniques such as one-shot encoding, label encoding or binary encoding.
    - b. Numerical variables can be normalized using techniques such as min-max scaling or standard scaling.
    - c. Dimensionality reduction techniques such as principal component analysis (PCA) can be used to reduce the number of features in the dataset.

- Classification:
    - a. The target variable to be predicted can be identified. For example, we can predict whether a student will pass or fail the final exam.
    - b. The dataset can be divided into training and testing sets.
    - c. Various classification algorithms such as logistic regression, decision tree or random forest can be used to train the model
    - d. The performance of the model can be evaluated using metrics such as accuracy, precision, recall, F1 score and ROC curve.

- Regularization:
    - a. It can be determined which variables are the most important when predicting the target variable. We can use feature selection techniques such as L1 regularization or recursive feature elimination.
    - b. Regularization techniques such as L1 or L2 regularization can be used to penalize the coefficients of the model and prevent overfitting.
    - c. The regularization parameter can be adjusted using cross-validation techniques such as k-fold cross-validation to improve the performance of the model.
